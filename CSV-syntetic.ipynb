# A1-synthetic.txt

### Read data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the data
df = pd.read_csv('A1-synthetic.txt' , sep='\t')
df_og = df.copy()
# Print the first 5 rows
df.head()
```

### Normalize data

```python
# Normalize the data to 0 - 1 of the first 9 columns
for column in df.columns[0:9]:
    df[column] = df[column] / df[column].max()

# Scale the z column to 0.1 - 0.9
df['z'] = df['z'] / df['z'].max() * 0.8 + 0.1

# Print the first 5 rows
df.head()
```

```python
# Output the data to a CSV file
df.to_csv('A1-synthetic-normalized.csv', index=False)
```

```python
# Unscale data
df['z'] = (df['z'] - 0.1) / 0.8 * df_og['z'].max()

# Unnormalize data
for column in df.columns[0:9]:
    df[column] = df[column] * df_og[column].max()

# Print the first 5 rows
df.head()
```
