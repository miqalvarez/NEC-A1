```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the data
df = pd.read_csv('A1-synthetic.txt' , sep='\t')

# Print the first 5 rows
df.head()

```

```python
# Normalize the data to 0 - 1 of the first 9 columns

df[0] = df[0] / df[0].max()
df[1] = df[1] / df[1].max()
df[2] = df[2] / df[2].max()
df[3] = df[3] / df[3].max()
df[5] = df[5] / df[5].max()
df[6] = df[6] / df[6].max()
df[7] = df[7] / df[7].max()
df[8] = df[8] / df[8].max()

# Scale the prediction column to 0.1 - 0.9
df[9] = df[9] * 0.8 + 0.1

# Print the first 5 rows
df.head()
```
